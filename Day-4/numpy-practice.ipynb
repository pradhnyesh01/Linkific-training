```python
import numpy as np
```

```python
a = np.arange(6)
print(f"The array shape is: {a.shape}")
print(a)
```

```python
M = np.array([[1, 2, 3],
              [4, 5, 6]])
print(f"The matrix shape is: {M.shape}")
print(M)
```

```python
C = np.arange(24). reshape(2, 3, 4)
print(f"The shape of C is: {C.shape}")
print(C)
```

```python
a = np.arange(10)**3
print(a)
print(a[9])
print(a[2:5])
print(a[-1])
print(a[::-1])
```

```python
b = np.arange(20).reshape(5,4)
print(b)
print(b[2,3])
print(b[0:3, 1])
print(b[:, 2:4])
```

```python
data = np.array([1, 2, 3])
ones = np.ones(3, dtype=int)
print(data+ones)
print(data-ones)
print(data*ones)
print(data/ones)
```

```python
a = np.arange(10) ** 2
print(a)
print(np.sum(a))
```

```python
b = np.array([[1, 1],
             [2, 2]])
print(np.sum(b, axis=0))
print(np.sum(b, axis=1))
```

```python
X = np.arange(10)
print(X.mean())
print(X.std())
```

```python
print(X * 2.5)
```

```python
A = np.arange(0, 40, 10). reshape((4,1))
print(A)
B = np.array([0, 1, 2])
print(B)
C = A + B
print(C)
```

```python
print(C.ravel()) # used for flattening
```

```python
a = np.arange(6)
row = a[np.newaxis, :]
col = a[:, np.newaxis]
print(row.shape, col.shape)
```

```python
rng = np.random.default_rng(seed=42)
R = rng.random((3,3))
print(R)
```

```python
rng = np.random.default_rng(seed=2026)
I = rng.integers(low=0, high=10, size=(4,5))
print(I)
```

```python
a = np.arange(12).reshape(3,4)
mask = a > 4
print(mask)
selected = a[mask]
print(selected)
```

```python
a = np.arange(10) ** 2
i = np.array([1, 1, 2, 4, 6])
print(a[i])
```

```python
a = np.arange(6)
b = np.arange(6, 10)
c = np.concatenate((a, b))
print(c)
```

```python
X = np.array([[1, 2],
              [3, 4]])
y = np.array([[5, 6]])
Z = np.concatenate((X, y), axis=0)
print(Z)
```

```python
n = np.array([10, 40, 30, 20, 50])
print(np.sort(n))
```

```python
D = np.arange(24).reshape((3,2,4))
print(D)
print(D.ndim)
print(D.size)
print(D.shape)
```

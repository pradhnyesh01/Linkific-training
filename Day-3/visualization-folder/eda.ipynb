```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./car_sales_data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
numeric_cols=df.select_dtypes(include='number').columns.tolist()
categorical_cols=df.select_dtypes(include='object').columns.tolist()
```

```python
X = df[numeric_cols[0]]
y = df[numeric_cols[1]]
```

```python
plt.figure()
plt.plot(df['Price'])
plt.title("Line plot")
plt.xlabel("Price")
plt.ylabel("Values")
plt.show()
```

```python
plt.figure()
plt.bar(range(len(X[:10])), X[:10])
plt.title("Bar Plot")
plt.xlabel("Index")
plt.ylabel(numeric_cols[0])
plt.show()
```

```python
plt.figure()
plt.scatter(df['Price'], df['Year of manufacture'])
plt.title("Scatter Plot")
plt.xlabel(numeric_cols[3])
plt.ylabel(numeric_cols[1])
plt.show()
```

```python
if categorical_cols:
    counts = df[categorical_cols[0]].value_counts().head(5)
    plt.figure()
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
    plt.title("Pie Chart")
    plt.show()
else:
    vals = X[:5]
    plt.figure()
    plt.pie(vals, labels=[f"Slice {i}" for i in range(len(vals))], autopct='%1.1f%%')
    plt.title("Pie Chart")
    plt.show()
```

```python
plt.figure()
plt.hist(X)
plt.title("Histogram")
plt.xlabel(numeric_cols[0])
plt.ylabel("Frequency")
plt.show()
```

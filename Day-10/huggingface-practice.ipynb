```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")

result = classifier("I love learning AI, it's amazing!")
print(result)
```

```python
texts = [
    "This movie was fantastic!",
    "The product is terrible.",
    "It was okay, not great."
]

results = classifier(texts)
for text, res in zip(texts, results):
    print(f"{text} → {res}")
```

```python
generator = pipeline("text-generation", model="gpt2")

output = generator(
    "Artificial Intelligence will",
    max_new_tokens=50,
    truncation=True,
    pad_token_id=50256
)

print(output)
```

```python
prompts = [
    "The future of AI is",
    "In 2050, humans will",
    "Machine learning helps in"
]

for prompt in prompts:
    result = generator(
        prompt,
        max_new_tokens=40,
        truncation=True,
        pad_token_id=50256
    )
    
    print(result[0]['generated_text'])
    print("-" * 50)
```

```python
qa = pipeline("question-answering")
context = """
Hugging Face is an AI company based in New York.
It develops tools for natural language processing.
"""

question = "What does Hugging Face do?"
```

```python
result = qa(question=question, context=context)
print(result)
```

```python
ner = pipeline("ner", grouped_entities=True)

text = "Elon Musk founded Tesla and lives in the USA."

result = ner(text)
print(result)
```

```python
translator = pipeline("translation_en_to_fr")

result = translator("Machine learning is very powerful.")
print(result)
```

```python
summarizer = pipeline("summarization")

text = """
Artificial Intelligence (AI) is one of the most transformative technologies of the 21st century. It enables machines to mimic human intelligence, including learning, reasoning, problem-solving, and decision-making. AI systems are powered by algorithms and vast amounts of data, allowing them to identify patterns and make predictions with high accuracy.
In healthcare, AI is used for disease diagnosis, drug discovery, and personalized treatment plans. In finance, it helps detect fraudulent transactions and optimize investment strategies. The transportation sector benefits from AI through autonomous vehicles and traffic management systems.
Despite its advantages, AI also raises concerns about job displacement, data privacy, and ethical decision-making. As AI continues to evolve, it is crucial to develop regulations and frameworks that ensure its responsible and fair use.
"""

summary = summarizer(
    text,
    max_length=80,
    min_length=30,
    do_sample=True,
    temperature=0.7
)
print(summary)
```

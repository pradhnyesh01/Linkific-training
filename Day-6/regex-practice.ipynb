```python
import re
```

```python
text = "Contact us at support@gmail.com or admin@yahoo.in"
emails = re.findall(r'\b[A-Za-z0-9._%+-]+@[A-Za-z]+\.[A-Za-z]{2,}\b',text)
print(emails)
```

```python
text = "Call me at +91-9876543210 or 9988776655"
phone_nums = re.findall(r'(?:\+91[\-\s]?)?[6-9]\d{9}', text)
print(phone_nums)
```

```python
text = "Dates: 12-05-2023, 2024/06/11, 15 Aug 2022"
dates = re.findall(r'\b\d{2}[-/]\d{2}[-/]\d{4}|\d{4}[-/]\d{2}[-/]\d{2}\b', text)
print(dates)
```

```python
text = "Hello!!! How are you?? #awesome"
clean = re.sub(r'[^A-Za-z0-9\s]', '', text)
print(clean)
```

```python
text = "This   is   messy   text"
clean = re.sub(r'\s+', ' ', text).strip()
print(clean)
```

```python
text = "Loving #AI and #MachineLearning"
hashtags = re.findall(r'#\w+', text)
print(hashtags)
```

```python
text = "Visit https://example.com"
urls = re.findall(r'https?://\S+', text)
print(urls)
```

```python
def is_valid_email(email):
    return bool(re.match(r'^[\w\.-]+@[\w\.-]+\.\w+$', email))
print(is_valid_email("abc@gmail.com"))
```

```python
text = "Price is 500 and discount is 20%"
nums = re.findall(r'\d+', text)
print(nums)
```

```python
text = "India is a Great Country"
words = re.findall(r'\b[A-Z][a-z]*\b', text)
print(words)
```

```python
text = "Meeting at 14:30 or 09:45"
times = re.findall(r'\b\d{2}:\d{2}\b', text)
print(times)
```

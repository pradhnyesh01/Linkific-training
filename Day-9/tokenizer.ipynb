```python
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Tokenization -> Split sentence into tokens
text = "I like watching thriller movies."
tokens = word_tokenize(text)
print(tokens)
```

```python
# Stopwords -> Common words like a, an, the, I, etc.
stop_words = set(stopwords.words('english'))
filtered = [w for w in tokens if w.lower() not in stop_words]
print(filtered)
```

```python
# Stemming -> Chops words to root form
ps = PorterStemmer()
print(ps.stem("running"))
```

```python
ps = PorterStemmer()

text = "I like watching thriller movies."
tokens = word_tokenize(text)

stemmed = [ps.stem(word) for word in tokens]

print(stemmed)
```

```python
# Lemmatization -> Context-aware root word
lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize("running", pos='v'))
```

```python
lemmatizer = WordNetLemmatizer()

lemmatized = [lemmatizer.lemmatize(word) for word in tokens]
print(lemmatized)
```

```python
# Bag of words -> Covert text to frequency vectors
corpus = ["I love space-movies", "Space movies are amazing"]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)

print(vectorizer.get_feature_names_out())
print(X.toarray())
```

```python
# TF-IDF -> Weights important words higher
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)

print(X.toarray())
```

```python
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z]', ' ', text)

    words = text.split()
    words = [w for w in words if w not in stopwords.words('english')]

    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(w) for w in words]

    return " ".join(words)

print(preprocess("I loved the movie! It was amazing."))
```
